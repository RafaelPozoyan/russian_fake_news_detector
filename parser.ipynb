```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://azbukamedia.com/arka-nezavisimosti/'
response = requests.get(url)
response.encoding = 'utf-8'
soup = BeautifulSoup(response.text, 'html.parser')

# Заголовок
title_tag = soup.find('h1')
title = title_tag.get_text(strip=True).lower() if title_tag else ''

# Находим блок после навигации (обычно это div с классом "entry-content", но бывает его нет)
# Если есть координатный навигационный блок, пропускаем его и берем только параграфы текста
main_text = []
# Попробуем самый универсальный способ — взять все <p> после первого h1 (но до следующего h2/h3)
if title_tag:
    # двигаемся по братьям после h1:
    for sib in title_tag.find_all_next():
        if sib.name == 'h2' or sib.name == 'h3':
            break
        if sib.name == 'p':
            pt = sib.get_text(strip=True).lower()
            if len(pt) >= 30:
                main_text.append(pt)
main_text = '\n'.join(main_text)

print('Заголовок:', title)
print('Текст (начало):', main_text[:400])

df = pd.DataFrame([{"title": title, "text": main_text}])
df.to_csv("data/azbuka_single_news.csv", index=False, encoding='utf-8-sig')
```

```python
# Получаем список всех div с текстом, ищем самый длинный
divs = soup.find_all('div')
max_txt = ''
for div in divs:
    txt = div.get_text(separator='\n', strip=True).lower()
    if len(txt) > len(max_txt):
        max_txt = txt

print("Самый длинный блок:\n", max_txt[:1500])
```

```python
st_accept = "text/html" # говорим веб-серверу, 
                        # что хотим получить html
# имитируем подключение через браузер Mozilla на macOS
st_useragent = "Mozilla/5.0 (Macintosh; Intel Mac OS X 12_3_1) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/15.4 Safari/605.1.15"
# формируем хеш заголовков
headers = {
   "Accept": st_accept,
   "User-Agent": st_useragent
}

# импортируем модуль
import requests

# отправляем запрос с заголовками по нужному адресу
req = requests.get("https://azbukamedia.com/arka-nezavisimosti/", headers)
# считываем текст HTML-документа
src = req.text
print(src)
```

```python
# инициализируем html-код страницы 
soup = BeautifulSoup(src, 'lxml')
# считываем заголовок страницы
title = soup.title.string
print(title)
```
